```python
import os 
import pandas as pd
import re 
import boto3
```

```python
all_data = pd.DataFrame()

for data in os.listdir(f'{os.getcwd()}/Hotels Information'):
    all_data = pd.concat([all_data, pd.read_csv(f'./Hotels Information/{data}')], ignore_index=True)

#all_data.fillna('Click Booking Link for More Info', inplace=True)

all_data.to_csv('All Hotels Information.csv')

all_data.head(5)
```

```python
all_data['Number of Reviews'] = all_data['Number of Reviews'].str[1:-1]

```

```python
all_data['Number of Reviews'] = all_data['Number of Reviews'].str.replace(',','')

all_data.head(5)
```

```python
all_data['Number of Reviews'] = pd.to_numeric(all_data['Number of Reviews'])  

all_data.head(20)
```

```python
all_data['Number of Reviews'] = all_data['Number of Reviews'].astype('Int64')

all_data.head()
```
